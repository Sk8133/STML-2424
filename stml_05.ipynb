```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
df=pd.read_csv('/content/archive (4).zip')
print(df.head())
features=df.iloc[:,:-1]
target=df.iloc[:,-1]
print("Features: ")
print(features.head())
print("\nTarget: ")
print(target.head())
for column in features.columns:
  plt.figure(figsize=(10,5))
  plt.scatter(features[column],target,color='orange')
  plt.xlabel(column)
  plt.ylabel("Target")
  plt.title(f"{column} vs Target")
  plt.show()
print(features.head())
```
